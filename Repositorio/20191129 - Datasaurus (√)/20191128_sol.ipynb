```python
df=pd.read_csv("DatasaurusDozen.tsv",sep="\t")
```

```python
df.head()
```

```python
index=1
plt.figure(figsize=(20,20))
for i in df["dataset"].unique():
    plt.subplot(4,4,index)
    x=df[df["dataset"]==i]["x"]
    y=df[df["dataset"]==i]["y"]
    plt.scatter(x,y,s=10)
    plt.title("dataset: "+i)
    index+=1
```

### Plot stats (using Pandas)

```python
for i in df["dataset"].unique():
    x=df[df["dataset"]==i]["x"]
    y=df[df["dataset"]==i]["y"]
    x_mean=np.mean(x)
    y_mean=np.mean(y)
    x_std=np.std(x)
    y_std=np.std(y)
    corr=np.corrcoef(x, y)[0][1]
    print("Dataset: %10s   X-Mean: %7.2f   Y-Mean: %7.2f   X_STD: %7.2f   Y_STD: %7.2f   PERASONS_CORR: %7.2f" % (i,x_mean,y_mean,x_std,y_std,corr))
```

### Plot datasets (using regular Python)

```python
dictionary={}
with open("DatasaurusDozen.tsv","r") as f:
    header=f.readline().split("\t")
    for row in f:
        key,x,y=row.split("\t")
        if key not in dictionary:
            dictionary[key]={"x":[float(x)],"y":[float(y)]}
        else:
            dictionary[key]["x"].append(float(x))
            dictionary[key]["y"].append(float(y))
```

```python
index=1
plt.figure(figsize=(20,20))
for i in dictionary.keys():
    plt.subplot(4,4,index)
    x=dictionary[i]["x"]
    y=dictionary[i]["y"]
    plt.scatter(x,y,s=10)
    plt.title("dataset: "+i)
    index+=1
```

### Plot stats (using regular Python)

```python
for i in dictionary.keys():
    x=dictionary[i]["x"]
    y=dictionary[i]["y"]
    x_mean=np.mean(x)
    y_mean=np.mean(y)
    x_std=np.std(x)
    y_std=np.std(y)
    corr=np.corrcoef(x, y)[0][1]
    print("Dataset: %10s   X-Mean: %7.2f   Y-Mean: %7.2f   X_STD: %7.2f   Y_STD: %7.2f   PERASONS_CORR: %7.2f" % (i,x_mean,y_mean,x_std,y_std,corr))
```
