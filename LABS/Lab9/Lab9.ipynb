### Lab 9.4. Plot 3D (1 pt.)

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x=np.array(np.linspace(0,2*np.pi,500))
y = x.copy()
xx, yy=np.meshgrid(x,y)
z = np.sin(xx)+ np.cos(yy)
surf = ax.plot_surface(xx,yy,z,cmap=cm.coolwarm)

# Añadir una leyende a la derecha de los diferentes valores en función del color
fig.colorbar(surf, shrink=0.85, aspect=5)

ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')

plt.show()
```

### Lab 9.5. Anscombe's quartet (1 pt.)

```python
import matplotlib.pyplot as plt
x = [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5]
y1 = [8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68]
y2 = [9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74]
y3 = [7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73]
x4 = [8, 8, 8, 8, 8, 8, 8, 19, 8, 8, 8]
y4 = [6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89]
ys= [y1,y2,y3,y4]
fig, axes = plt.subplots(2,2,figsize=(10,10))
i,w,v,f=0,1,0,0
romanos_indi='I'
romanos_cin='V'
for y in ys:
    if v >1:
        v=0
        i+=1
    if w>3:
        w=1
        f+=1
    if i==1 and v==1:
        axes[i,v].plot(x4,y,'ks')
    else:
        axes[i,v].plot(x,y,'ks')

    axes[i,v].text(17,12,romanos_indi*w+romanos_cin*f,size=20,color='b')
    axes[i,v].set_xlim(2.5,20)
    axes[i,v].set_ylim(2,14)
    v+=1
    w+=1  
plt.show()
```

```python
for i in ys:
    #Media
    mn=np.mean(i)
    
    #Desviación estandar
    ds=np.std(i)
    
    #Coeficiente de correlación
    if i !=y4:
        coor=np.corrcoef(x,i)[0][1]
    else:
        coor=np.corrcoef(x4,i)[0][1]  
    #Lo que nos quiere decir que sea positvo es que cuanto mayor sea el valor de x mayor es valor de y
    print(f'La media: {mn:0.2f},desviación típica {ds:0.2f} y coeficiente de correlación {coor:0.2f}')
```

## Lab 9.6 BiciMad Data (3 pt.)

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
bicimad =pd.read_csv('bicimad.csv','r',delimiter=';')
plt.xlabel('Groups of age')
plt.ylabel('Number of travelers')
plt.hist(bicimad['ageRange'],bins=[0,1,2,3,4,5,6,7],rwidth=0.5,align='left')
legnda=['0: No se ha podidio determinar el rango de edad del usuario','1: El usuario tiene ente 0 y 16 años','2: El usuario tiene ente 17 y 18 años','3: El usuario tiene ente 19 y 26 años','4: El usuario tiene ente 27 y 40 años','5: El usuario tiene ente 41 y 65 años','6: El usuario tiene ente 66 años o más']
v=[]
for i in legnda:
    v.append(mpatches.Patch(label=i))
plt.legend(handles=v,bbox_to_anchor=(1.05,1))
plt.show()
```
