```python
def check_coincidences(guess,secret):
    out=[]
    for i in range(len(guess)):
        if guess[i]==secret[i]:
            out.append(1)
        else:
            out.append(0)
    number_of_coincidences=sum(out)
    return number_of_coincidences
```
